```python
gestures = [
    
#     'Fist',
#     'Five',
#     'Four',
#     'Heart_1',
#     'Heart_2',
#     'Heart_3',
#     'Heart_single',
#     'Honour',
#     'ILY',
#     'Insult',
#     'Nine',
#     'OK',
#     'One',
#     'Palm_up',
#     'Prayer',
    'Rock',
    'Seven',
    'Six',
    'Three',
    'Thumb_down',
    'Thumb_up',
    'Two'
]
```

```python
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
import os

for name in gestures:
    path = 'Gestures_all//' + name

    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.15,
        height_shift_range=0.15,
        zoom_range=0.15,
        shear_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')

    dirs = os.listdir(path)
    length = len(dirs)
    print(length)
    for filename in dirs:
        img = load_img(path + "//{}".format(filename))
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape) #datagen.flow要求rank为4
        datagen.fit(x)
        prefix = filename.split('.')[0]

        print(str(length) + '(' + prefix + ')')
        length = length - 1

        counter = 0
        for batch in datagen.flow(x, batch_size=4 , save_to_dir=path, save_prefix=prefix, save_format='jpg'):
            counter += 1
            if counter > 20:
                break  # 否则生成器会退出循环
```
